```python
#https://stackoverflow.com/questions/50534778/python-extract-keywords-row-by-row-from-csv
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


stop_words = stopwords.words('english')
```

```python
def get_keywords(row):
    some_text = row['Notes']
    lowered = some_text.lower()
    tokens = nltk.tokenize.word_tokenize(lowered)
    keywords = [keyword for keyword in tokens if keyword.isalpha() and not keyword in stop_words]
    keywords_string = ','.join(keywords)
    return keywords_string
```

```python
df_2013_14['keywords'] = df_2013_14.apply(get_keywords, axis=1)
df_2013_14.head()
```
