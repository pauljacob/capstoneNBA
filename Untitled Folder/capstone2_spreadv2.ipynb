```python
def american_odds_to_implied_prob(df_, col_):
    ml_neg = df_[df_[col_] < 0][col_]
    favorite = (-1 * ml_neg) / (100 + (-1) * ml_neg)
    
    ml_pos = df_[df_[col_] > 0][col_]
    underdog = 100 / (ml_pos + 100)
    
    df_ = pd.concat([favorite, underdog], axis=0)
    return df_.sort_index()

```

```python
df['price1_implied_prob'] = american_odds_to_implied_prob(df, 'price1')
df['price2_implied_prob'] = american_odds_to_implied_prob(df, 'price2')
```
