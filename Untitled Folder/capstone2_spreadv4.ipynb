visitor = np.array(df_games.game_id.unique().shape[0])
home = np.array(df_games.game_id.unique().shape[0])

#collect the scores
for game_id_ in df_games.game_id.unique():
    
    #get visitor and home team single game stats
    df_games[df_games['game_id'] == game_id_]
    
    home = home + (df_games[df_games['game_id'] == game_id_][df_games['is_home'] == 't'][['game_id','pts']])
    
    visitor = visitor + (df_games[df_games['game_id'] == game_id_][df_games['is_home'] == 'f'][['game_id','pts']])
 
    
    

```python
#visitor.shape
```

```python
#columns of interest in df_games are ['game_id', 'is_home', 'pts']

df_games[['game_id', 'is_home', 'pts']]

#df_games_home = df_games[['game_id', 'is_home', 'pts']][df_games['is_home'] == 't']
df_games_home = df_games[['game_id', 'pts']][df_games['is_home'] == 't'].rename(columns={'pts':'pts_home'})
df_games_home
```

```python
#df_games_not_home = df_games[['game_id', 'is_home', 'pts']][df_games['is_home'] == 'f']
df_games_not_home = df_games[['game_id', 'pts']][df_games['is_home'] == 'f'].rename(columns={'pts':'pts_not_home'})
df_games_not_home
```

```python
df_games_pts = df_games_not_home.merge(df_games_home)
df_games_pts
```

```python
#actual point difference by game_id
df_games_pts['pts_diff'] = - (df_games_pts['pts_not_home'] - df_games_pts['pts_home'])
df_games_pts
```
